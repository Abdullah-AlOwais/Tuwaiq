## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
df
```

```python
df.info()
```

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
df.dropna(inplace=True)
df.info()
```

```python
df.describe()
```

### Identify and analyze patterns in the data.

```python
# correlation matrix
numerical_columns = df[['TrafficIndexLive','JamsCount','JamsDelay','JamsLength','TrafficIndexWeekAgo','TravelTimeHistoric','TravelTimeLive']]
correlation_matrix = numerical_columns.corr()
print(correlation_matrix)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['TravelTimeLive'], kde=True, bins=10)
sns.histplot(df['TravelTimeHistoric'], kde=True, bins=10)
plt.title('Distribution of feature')
plt.xlabel('TrafficIndexLive')
plt.ylabel('TravelTimeHistoric')
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
# Create a scatter plot to visualize the relationship between 'JamsDelay' and 'JamsCount'.
plt.scatter(data=df,x='JamsDelay',y='JamsCount')
plt.title('The Correlation between JamsDelay and JamsCount')
plt.xlabel('JamsDelay')
plt.ylabel('JamsCount')
plt.show()
```
